Testing indexing logic

```python
import torch
torch.set_printoptions(threshold=1000)
```

```python
N = 1
SEQ = 1024
WIDTH = 1536
NUM_HEADS = 16
HEAD_DIM = WIDTH // NUM_HEADS
print(f"N: {N}, SEQ: {SEQ}, WIDTH: {WIDTH}, NUM_HEADS: {NUM_HEADS}, HEAD_DIM: {HEAD_DIM}")
```

```python
torch.manual_seed(42)
x = torch.rand((N,NUM_HEADS,SEQ,HEAD_DIM))
y = x.permute((0,2,1,3))
y = y.reshape((N,SEQ,NUM_HEADS*HEAD_DIM))
```

```python
print(x[0,0,0,:]) # This is 0th head embeding of the 0th sequence
print(x[0,1,0,:])
```

```python
torch.set_printoptions(threshold=10000) # Set a high threshold

print(y[0,0,:96]) # This is all the heads in the 0th sequence. WIDTH // HEAD_IDX = 96 each
print(y[0,0,96:192]) # This is all the heads in the 0th sequence. WIDTH // HEAD_IDX = 96 each
```

```python
z = y.reshape((N,SEQ,NUM_HEADS,HEAD_DIM)).permute((0,2,1,3))
(x-z).mean()
```
